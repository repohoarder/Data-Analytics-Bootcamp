```python
model3hs=pd.get_dummies(data=model2hs, columns=['WARD'])
model3hs.head()
```

```python
X=model3hs.drop('PRICE', axis=1)
```

```python
lm.fit(X, model3hs.PRICE)
```

```python
pd.DataFrame(list(zip(X.columns, lm.coef_)), columns = ['features', 'estimatedCoefficients'])
```

```python
#test_size default = 0.20
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(
    X, model2hs.PRICE, test_size=0.20, random_state = 5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lm=LinearRegression()
lm.fit(X_train, Y_train)
pred_train = lm.predict(X_train)
pred_test = lm.predict(X_test)
```

```python
print ('Fit a model X_train, and calculate MSE with Y_train:', np.mean((Y_train - lm.predict(X_train)) ** 2))
print ('Fit a model X_train, and calculate MSE with X_test, Y_test:', np.mean((Y_test - lm.predict(X_test)) ** 2))
```

```python
hs2=model2hs
```

```python
#extract target variable
#make copy of 'WARD' column
y = hs2['WARD']
```

```python
hs2.dtypes
```

```python
X = hs2.drop(['WARD'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=90)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
#develop model using training data
#defining arguments in the model can help prevent overfitting
model.fit(X_train, y_train)
```

```python
#run the predictions on the test data
y_predict = model.predict(X_test)
```

```python
#check the accuracy of model
accuracy_score(y_test, y_predict)
```

```python
#look at true and false predictions
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    )
```
