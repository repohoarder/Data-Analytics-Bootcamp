```python
meanP=housing['PRICE'].mean()
stdP=housing['PRICE'].std()
topP=meanP + stdP *1.96
botP=meanP -stdP *1.96
```

```python
housingsmall_P2 = pd.DataFrame(data = housingsmall, columns=['PRICE','GBA','ZIPCODE','SALEDATE','BEDRM','BATHRM','CITY','HF_BATHRM','AYB','WARD','LIVING_GBA'])
housingsmall_P2.head()
```

```python
sns.lmplot(x='ZIPCODE', y='PRICE', data=housingsmall_P2)
```

```python
sns.lmplot(x='ZIPCODE', y='PRICE', data=housingsmall_P2,
           fit_reg=False, #remove regression line
           hue='BEDRM')
```

```python
sns.lmplot(x='GBA', y='PRICE', data=housingsmall_P2,
           fit_reg=False, #remove regression line
           hue='BEDRM')
```

```python
plt.subplots(figsize=(13,7))

sns.violinplot(x='BEDRM', y='PRICE', data=housingsmall_P2)
```
